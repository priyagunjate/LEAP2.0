```python
#boston data housing datasets (inbuilt datsets)
from sklearn.datasets import load_boston
```

```python
boston=load_boston()

# Dimension data
print(boston.data.shape)
```

```python
#features 
print(boston.feature_names)
```

```python
print(boston.DESCR)
```

```python
boston.target
```

```python
# Write dataframe
import pandas as pd
bos= pd.DataFrame(boston.data)
bos.head()
```

```python
bos['PRICE']=boston.target
bos
```

```python
# X  ,Y

X= bos.drop('PRICE',axis=1)  
Y=bos['PRICE']
```

```python
#X=X[1:]
#X
```

# why will you train test splitting??

```python
# Train  cv test splitting
# performance metrices will high

Datasets : x, y cross_validation , y_test (70 10 20)
    80 - 20  opr 70 -30 %

    
cross_validation - k-fold cross validation { k-4}

D0, D1, D2, D3 cv=D3
D0,D1,D3,D2 cv=d2
```

```python
from sklearn.model_selection  import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.3,random_state=1)
```

```python
print('Traing',X_train.shape)
print('Testing datasets',X_test.shape)
print('Y training',Y_train.shape)
print('Y_test',Y_test.shape)
```

```python
# Linear regression

from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,Y_train) # Fitting 
```

```python
y_pred= lm.predict(X_test)
```

```python
y_pred   # Comapre Y_test ... error accuracy performace metricss
```
