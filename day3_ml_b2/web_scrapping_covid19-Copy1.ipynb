# Web API 

```python
url='https://www.mohfw.gov.in/'
web_content=requests.get(url).content

soup=BeautifulSoup(web_content,"html.parser")
print(soup)
extract_contents=lambda row : [x.text.replace('\n', '') for x in row]
stats=[]
all_rows= soup.find_all('tr')
for row in tqdm(all_rows):
    stat= extract_contents(row.find_all('td'))
    if len(stat) == 6:
        stats.append(stat)
```

```python
stats
```

```python
new_cols=['sr.No','States','confirmed','recovered','deaths','Totol cases']
state_df=pd.DataFrame(data=stats,columns=new_cols)
state_df
```

```python
state_df.isnull().sum()
```

```python
state_df=state_df[0:35]
state_df.tail()
```

```python
type(state_df['confirmed'][1])
```

```python
confirmed=[]
recovered=[]
deaths=[]
total_case=[]
for i in range(len(state_df)):
    confirmed.append(int(state_df['confirmed'][i]))
    recovered.append(int(state_df['recovered'][i]))
    deaths.append(int(state_df['deaths'][i]))
    total_case.append(int(state_df['Totol cases'][i]))
```

```python
state_df['confirmed']=confirmed
state_df['recovered']=recovered
state_df['deaths']=deaths
state_df['Totol cases']=total_case
```

# Plotting Donut Chart

```python
group_size=[sum(state_df['confirmed']),
               sum(state_df['recovered']),
                  sum(state_df['deaths']),
                     sum(state_df['Totol cases'])]
```

```python
group_labels=['confimred\n'+str(sum(state_df['confirmed'])),
             'recovered\n'+str(sum(state_df['recovered'])),
            'deaths\n'+str(sum(state_df['deaths'])) ,
            'total_cases\n'+str(sum(state_df['Totol cases'])) ]
```

```python
colors=['skyblue','yellowgreen','red','tomato']
```

```python
plt.figure(figsize=(5,5))
plt.pie(group_size,labels=group_labels,colors=colors)
plt.show()
```
